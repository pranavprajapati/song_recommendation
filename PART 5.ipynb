```python
%matplotlib inline

import pandas as pd
from sklearn.datasets import load_digits
import hypertools as hyp
```

```python
df = pd.read_csv('main_cluster.csv')
# Subset the column of the clustering result
target = df['main_clustering.clustering']
df = df.drop('main_clustering.clustering', axis=1)
group = target.astype('str')
```

```python
hyp.plot(df, '.', group=group, model='TSNE', ndims=2, legend=[1,2,3],
         title = "2D Representation of the Main Cluster using t-SNE")
```

```python
hyp.plot(df, '.', group=group, model='PCA', ndims=2, legend=[1,2,3], title = "2D Representation of the Main Cluster using PCA")
```

```python
hyp.plot(df, '.', group=group, model='TSNE', ndims=3, legend=[1,2,3], title = "3D Representation of the Main Cluster using t-SNE")
```
