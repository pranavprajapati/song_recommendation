```python
import pandas as pd


train_data = pd.read_excel('Train.xlsx', encoding='ISO-8859-1')
test_data = pd.read_csv('2016.csv', encoding='ISO-8859-1')
```

```python
train_data
```

```python
test_data
```

```python
train_data.set_index("name", inplace = True)
```

```python
ignore = (['artist', 
           'release_date', 
           'length', 
           'album',  
           'time_signature',
          ])


clean_test_data = test_data.drop(ignore, axis=1)
```

```python
test_data.set_index("name", inplace = True)
```

```python
train_data_x = train_data.drop('class', axis=1)
train_data_y = train_data[['class']]
```

```python
from sklearn import decomposition
from scipy.sparse import csr_matrix
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(style='white')
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
X_scaled = StandardScaler().fit_transform(train_data_x)

pca = decomposition.PCA().fit(X_scaled)

import numpy as np
plt.figure(figsize=(10,7))
plt.plot(np.cumsum(pca.explained_variance_ratio_), color='k', lw=2)
plt.xlabel('Number of components')
plt.ylabel('Total explained variance')
plt.xlim(0, 12)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.axvline(6, c='b') 
plt.axhline(0.96, c='r')#obtain a 96% total variance
plt.show();
```

```python
pca1 = decomposition.PCA(n_components=6)
X_pca = pca1.fit_transform(X_scaled)
```

```python
X_pca.shape
```

```python
train_data_song_names = train_data.index.values
```

```python
test_data_song_names = clean_test_data.index.values
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

v = TfidfVectorizer(sublinear_tf=True, ngram_range=(1, 6), max_features=10000)
X_names_sparse = v.fit_transform(train_data_song_names)
X_names_sparse.shape
```

```python
from scipy.sparse import csr_matrix, hstack

X_train_last = csr_matrix(hstack([X_pca, X_names_sparse]))
```

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
# kNN first
from sklearn.neighbors import KNeighborsClassifier

knn_params = {'n_neighbors': range(1, 10)}
knn = KNeighborsClassifier(n_jobs=-1)

knn_grid = GridSearchCV(knn, knn_params, cv=skf, n_jobs=-1, verbose=True)
knn_grid.fit(X_train_last, train_data_y)
knn_grid.best_params_, knn_grid.best_score_
```

```python
# Decision Trees second
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree_params = {'max_depth': range(1,8), 'max_features': range(2,8)}
tree_grid = GridSearchCV(tree, tree_params, cv=skf, n_jobs=-1, verbose=True)
tree_grid.fit(X_train_last, train_data_y)
tree_grid.best_estimator_, tree_grid.best_score_
```

```python
# Random Forests third
from sklearn.ensemble import RandomForestClassifier
parameters = {'max_features': [1, 2, 8], 'min_samples_leaf': [1, 3, 6], 'max_depth': [1, 9]}
rfc = RandomForestClassifier(n_estimators=100, random_state=42, 
                             n_jobs=-1, oob_score=True)
randomforestgrid = GridSearchCV(rfc, parameters, n_jobs=-1, cv=skf, verbose=1)
randomforestgrid.fit(X_train_last, train_data_y)
randomforestgrid.best_estimator_, randomforestgrid.best_score_
```

```python
test_data_x = clean_test_data.drop('class', axis=1)
test_data_y = clean_test_data[['class']]

```

```python
# Make predictions
knn_grid.best_estimator_.fit(X_train_last, train_data_y)
```

```python
X_test_scaled = StandardScaler().fit_transform(test_data_x)

#pca = decomposition.PCA().fit(X_test_scaled)
X_test_pca = pca1.fit_transform(X_test_scaled)
```

```python
X_test_pca = pca1.transform(X_test_scaled)
```

```python
X_test_names = v.transform(test_data_song_names)
```

```python
X_test_last = csr_matrix(hstack([X_test_pca, X_test_names]))
```

```python
X_test_last.shape
```

```python
y_pred_class = knn_grid.best_estimator_.predict(X_test_last)
```

```python
clean_test_data['class'] = y_pred_class
```
